```python
import smtplib
import ssl
from email.message import EmailMessage

EMAIL = "hiteshdhaked925@gmail.com"
APP_PASSWORD = "XXXXXXXXXXX"
RECEIVER = "hiteshdhaked925@gmail.com" #Enter recievers Email

msg = EmailMessage()
msg["From"]= EMAIL
msg["To"]=RECEIVER
msg["Subj"]="Hello For Python......"
msg.set_content("This mail was shared by python....")
context = ssl.create_default_context()
with smtplib.SMTP_SSL("smtp.gmail.com",465,context=context) as server:
    server.login(EMAIL, APP_PASSWORD)
    server.send_message(msg)
```
