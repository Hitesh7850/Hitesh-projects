```python
menu = [
    {"ID": "102", "Name": "Burger", "Price": 150},
    {"ID": "206", "Name": "Pizza", "Price": 280},
    {"ID": "332", "Name": "Pasta", "Price": 150},
    {"ID": "401", "Name": "Sandwich", "Price": 120},
    {"ID": "510", "Name": "Cold Drink", "Price": 80}
]

order_list = []

while True:
    print("====Online Food Ordering System====")
    print("1. View Menu")
    print("2. Place Order")
    print("3. View Bill")
    print("4. Exit")

    choice = input("Enter your choice : ")

    if choice == "1":
        print("====Menu====")
        for item in menu:
            print(f"ID: {item['ID']} | {item['Name']} | {item['Price']}")

    elif choice == "2":
        item_id = input("Enter Item ID to order: ")
        found = False
        for item in menu:
            if item["ID"] == item_id:
                order_list.append(item)
                print(f"{item['Name']} added to your order!")
                found = True
                break
        if not found:
            print("Invalid Item ID. Please order again.")

    elif choice == "3":
        if not order_list:
            print("No items ordered yet!")
        else:
            print("==== Your Order ====")
            total = 0
            for item in order_list:
                print(f"{item['Name']} - {item['Price']}")
                total += item["Price"]
            print(f"Total Bill: {total}")

    elif choice == "4":
        print("Thank you for ordering! Enjoy your meal üçΩÔ∏è")
        break

    else:
        print("Invalid choice. Please try again.")
```
