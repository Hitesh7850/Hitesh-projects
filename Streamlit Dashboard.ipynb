```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

st.title("Streamlit app for VGU ")
st.text("Welcome to our Dashboard")
st.header("I am a header")
st.write("You can write", 10+5)

name=st.text_input("Enter your name :")
age=st.number_input("Enter your age :")
st.write("Your name is :",name , "Your age is :",age)
st.selectbox("Enter your course :",["BCA", "B.Tech", "MCA"])
if st.button("Click Me"):
    st.success("Clicked Button")

data = {"Name":["Tom","Jack","HITESH"],"Marks": [10,20,40]}
df = pd.DataFrame(data)

st.dataframe(df)

data = pd.DataFrame({
    "Marks" : [10,20,20]
})

st.line_chart(data)
st.bar_chart(data)

subject = ["Python","C++","Java"]
marks = [20,10,15]

fig , Ad =plt.subplots()
Ad.pie(marks, labels=subject)
st.pyplot(fig)
```
